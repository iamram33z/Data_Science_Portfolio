# Data Clean & Pre-process

```python
import os

#Update the Database with relative paths
current_dir = os.getcwd()
task03_dir = os.path.dirname(current_dir)
database_dir = os.path.join(task03_dir, 'Database')
uncleaned_data_dir = os.path.join(database_dir, 'Uncleaned Data')
cleaned_data_dir = os.path.join(database_dir, 'Cleaned Data')

#Get the disaster data file path
disaster_data_path = os.path.join(uncleaned_data_dir, 'Flood Data','di_report-srilanka.csv')

#Get the map folders path
SriLankaMap_dir = os.path.join(uncleaned_data_dir, 'Sri Lanka')

DistrictsMap_dir = os.path.join(SriLankaMap_dir, 'Districts')
ProvincesMap_dir = os.path.join(SriLankaMap_dir, 'Provinces')
DivisionsMap_dir = os.path.join(SriLankaMap_dir, 'Divisions')

#Get the climate data folder path
climate_data_path = os.path.join(uncleaned_data_dir, 'Climatic Data')

#Get the Climatic Data Files
seasonal_avg_surf_air_temperature_data_path = os.path.join(climate_data_path, 'tas_c_seasonal.xlsx')
seasonal_avg_rainfall_data_path = os.path.join(climate_data_path, 'pr_mm_seasonal.xlsx')
```

```python
import pandas as pd

#Read the disaster data file
disaster_data = pd.read_csv(disaster_data_path)
disaster_data
```

```python
#Drop columns that are not required
disaster_data.drop(['Unnamed: 25', 'Unnamed: 26', 'Unnamed: 27','Unnamed: 28'], axis=1, inplace=True)
disaster_data
```

```python
#Count the unique values for 'Event' column
disaster_data['Event'].value_counts()
```

```python
#Get only the rows that have 'FLOOD',''URBAN FLOOD','CYCLONE & FLOOD',''FLASH FLOOD' in the 'Event' column
disaster_data = disaster_data[disaster_data['Event'].isin(['FLOOD','URBAN FLOOD','CYCLONE & FLOOD','FLASH FLOOD'])]
disaster_data
```

```python
Flood_Data = disaster_data.copy()

#Replace the 'URBAN FLOOD','CYCLONE & FLOOD',''FLASH FLOOD' in the 'Event' column with 'FLOOD'
Flood_Data['Event'] = Flood_Data['Event'].replace(['URBAN FLOOD','CYCLONE & FLOOD','FLASH FLOOD'],'FLOOD')
Flood_Data
```

```python
#Check for null values
Flood_Data.isnull().sum()
```

```python
#Remove 'Source' column
Flood_Data.drop(['Source'], axis=1, inplace=True)

#Remove 'Location' column
Flood_Data.drop(['Location'], axis=1, inplace=True)

#Remove Null values in 'Code Commune' column
Flood_Data.dropna(subset=['Code Commune'], inplace=True)

#Replace the null values in the 'Deaths','Injured','Missing','Houses Destroyed'with 0
Flood_Data['Deaths'] = Flood_Data['Deaths'].fillna(0)
Flood_Data['Injured'] = Flood_Data['Injured'].fillna(0)
Flood_Data['Missing'] = Flood_Data['Missing'].fillna(0)
Flood_Data['Houses Destroyed'] = Flood_Data['Houses Destroyed'].fillna(0)
Flood_Data

#Rename Code Region column with Code Province
Flood_Data.rename(columns={'Code Region':'Code Province'}, inplace=True)

#Rename Code Commune with Code Division
Flood_Data.rename(columns={'Code Commune':'Code Division'}, inplace=True)
```

```python
Flood_Data
```

```python
#seperate the day, month and year from the 'Date (YMD)' as string columns
Flood_Data['Year'] = Flood_Data['Date (YMD)'].str.split('/', expand=True)[2]
Flood_Data['Month'] = Flood_Data['Date (YMD)'].str.split('/', expand=True)[1]
Flood_Data['Day'] = Flood_Data['Date (YMD)'].str.split('/', expand=True)[0]

#Remove the 'Date (YMD)' column
Flood_Data.drop(['Date (YMD)'], axis=1, inplace=True)

#Remove leading 0 in 'Day' column
Flood_Data['Day'] = Flood_Data['Day'].str.lstrip('0')

#Remove leading 0 in 'Month' column
Flood_Data['Month'] = Flood_Data['Month'].str.lstrip('0')

#Replace 0 in 'Day' column with 1
Flood_Data['Day'] = Flood_Data['Day'].replace(['0'],'1')

#Replace 0 in 'Month' column with 1
Flood_Data['Month'] = Flood_Data['Month'].replace(['0'],'1')

#Set the 'Month' column to the relvant quartile month
def set_quartile_month(month):
    if month in ['1', '2', '3']:
        return '1'
    elif month in ['4', '5', '6']:
        return '4'
    elif month in ['7', '8', '9']:
        return '7'
    elif month in ['10', '11', '12']:
        return '10'
    else:
        return month

Flood_Data['Month'] = Flood_Data['Month'].apply(set_quartile_month)

#Set 'Month' and 'Year' as Season with MM-YYYY format
Flood_Data['Season'] = Flood_Data['Month'] + '-' + Flood_Data['Year']

#Remove 'Day', 'Month' and 'Year' columns
Flood_Data.drop(['Day', 'Month', 'Year'], axis=1, inplace=True)

Flood_Data
```

```python
#Sort by Season
Flood_Data.sort_values(by=['Season'], inplace=True)
Flood_Data
```

```python
#Read the seasonal average surface air temperature data
seasonal_avg_surf_air_temperature_data = pd.read_excel(seasonal_avg_surf_air_temperature_data_path)

#Get as MM-YYYY format from the 'Season' column
seasonal_avg_surf_air_temperature_data['Season'] = seasonal_avg_surf_air_temperature_data['Season'].dt.strftime('%m-%Y')

#Remove leading 0 in 'Season' column
seasonal_avg_surf_air_temperature_data['Season'] = seasonal_avg_surf_air_temperature_data['Season'].str.lstrip('0')

#Set 'Season' as Season with MM-YYYY format
seasonal_avg_surf_air_temperature_data['Season'] = seasonal_avg_surf_air_temperature_data['Season'].apply(set_quartile_month)

seasonal_avg_surf_air_temperature_data
```

```python
#Read the seasonal average rainfall data
seasonal_avg_rainfall_data = pd.read_excel(seasonal_avg_rainfall_data_path)

#Get as MM-YYYY format from the 'Season' column
seasonal_avg_rainfall_data['Season'] = seasonal_avg_rainfall_data['Season'].dt.strftime('%m-%Y')

#Remove leading 0 in 'Season' column
seasonal_avg_rainfall_data['Season'] = seasonal_avg_rainfall_data['Season'].str.lstrip('0')

#Set 'Season' as Season with MM-YYYY format
seasonal_avg_rainfall_data['Season'] = seasonal_avg_rainfall_data['Season'].apply(set_quartile_month)

seasonal_avg_rainfall_data
```

```python
#Left Join the Flood_Data with the seasonal_avg_surf_air_temperature_data and seasonal_avg_rainfall_data
Flood_Data = pd.merge(Flood_Data, seasonal_avg_surf_air_temperature_data, how='left', on='Season')
Flood_Data = pd.merge(Flood_Data, seasonal_avg_rainfall_data, how='left', on='Season')

#Remove 'Serial' and 'Event' columns
Flood_Data.drop(['Serial', 'Event'], axis=1, inplace=True)

Flood_Data
```

```python
#Save the cleaned data to a csv file
Flood_Data.to_csv(os.path.join(cleaned_data_dir, 'Flood Data','Flood_Data.csv'), index=False)
```
